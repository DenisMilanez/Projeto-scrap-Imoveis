```python
url = 'https://www.dfimoveis.com.br/venda/df/brasilia/noroeste/imoveis?&ordenamento=mais-recente'

def response_parser(url, headers):
    try:
        response = requests.get(url, headers=headers)

        # Verificando a requisição
        if response.status_code == 200:
            print(response)
            html_content = response.text
            soup = BeautifulSoup(html_content, 'html.parser')
            return soup
        else:
            print('Não foi possível acessar a página')
            return None
    except:
        print('erro')
        return None
```

```python
def calcula_paginas(soup):
    try:
        # Obtem valores e realiza calcula o número de páginas
        imoveis_por_pagina = int(soup.find('input', {'id': 'hidden-quantidade-de-imoveis-por-pagina'})['value'])
        imoveis_encontrados = int(soup.find('input', {'id': 'hidden-quantidade-de-imoveis-encontrados'})['value'])

        # soma imoveis_por_pagina - 1 para obter para obter número arredondado. 
        numero_paginas = int((imoveis_encontrados + imoveis_por_pagina - 1) // imoveis_por_pagina)
    except:
        print('Não foi possível calcular o número d epáginas')
        numero_paginas = None
    finally:
        print(f'Total de Imóveis: {imoveis_encontrados}. Imóveis por página: {imoveis_por_pagina}.\nNúmero de Páginas: {numero_paginas} ')
        return numero_paginas
```

```python
def scrap_lista(soup):
    dados_anuncio = [] # dicionario para armazenar dados

    try:
        # Isolando a div com a lista dos anúncios
        container_lista = soup.find('div', {'id': 'resultadoDaBuscaDeImoveis'})

        # Selecionando todos os elementos <a>
        itens_lista = container_lista.find_all('a')

        # Percorrendo e extraindo dados
        for item in itens_lista:
            # Link do anúncio
            try:
                url_anuncio = item['href']
            except:
                url_anuncio = None
            print(f'url_anuncio: {url_anuncio}')

            # Imagem do anúncio
            try:
                img_anuncio = item.find('div', class_='new-pic').find('source')['srcset']
            except:
                img_anuncio = None
            print(f'img_anuncio: {img_anuncio}')

            # Categoria do Anúncio
            try:
                categoria_anuncio = item.find('div', class_=re.compile(r'new-texts')).text.strip()
            except:
                categoria_anuncio = None
            print(f'categoria_anuncio: {categoria_anuncio}')

            # Titulo endereço
            try:
                titulo_endereco = item.find('div', 'new-head').find('h2', 'new-title phrase').text.strip()
            except:
                titulo_endereco = None
            print(f'titulo_endereco: {titulo_endereco}')

            # Subtitulo endereço
            try:
                subtitulo_endereco = item.find('div', 'new-head').find('h3', 'new-subtitle phrase').text.strip()
            except:
                subtitulo_endereco = None
            print(f'subtitulo_endereco: {subtitulo_endereco}')

            # Subtitulo frase
            try:
                subtitulo_frase = item.find('div', 'new-info').find('h3', 'new-simple phrase').text.strip()
            except:
                subtitulo_frase = None
            print(f'subtitulo_frase: {subtitulo_frase}')

            # Preço do anúncio
            try:
                preco = item.find('div', class_='new-price').find('h4').find('span').text.strip()
            except:
                preco = None
            print(f'preco: {preco}')

            # Area privativa:
            try:
                area_privativa = item.find('li', class_= 'm-area').text.strip()
            except:
                area_privativa = None
            print(f'area_privativa: {area_privativa}')

            # Lista detalhes
            try:
                ul = item.find('ul', class_= 'new-details-ul')
                lista_detalhes = [li.text.strip() for li in ul.find_all('li')]
                lista_resultado = ', '.join(lista_detalhes)
            except:
                lista_resultado = None
            print(f'lista_detalhes: {lista_resultado}')

            # Imagem do anunciante
            try:
                img_anunciante = item.find('div', class_='new-anunciante').find('source')['srcset']
            except:
                img_anunciante = None
            print(f'img_anunciante: {img_anunciante}')

            # Alt nome do anunciante
            try:
                nome_anunciante = item.find('div', class_='new-anunciante').find('img')['alt']
            except:
                nome_anunciante = None
            print(f'nome_anunciante: {nome_anunciante}')

            # Creci
            try:
                creci = item.find('div', class_='new-anunciante').find('p').text.strip()
            except:
                creci = None
            print(f'creci: {creci}')

            # Pagina do anúncio
            try:
                pagina = soup.find('input', {'id': 'hidden-pagina'})['value']
            except:
                pagina = None
            print(f'pagina: {pagina}')

            # cria um dicionário com a lista de dados
            dados = {
                'url_anuncio': url_anuncio,
                'img_anuncio': img_anuncio,
                'categoria_anuncio': categoria_anuncio,
                'titulo_endereco': titulo_endereco,
                'subtitulo_endereco': subtitulo_endereco,
                'subtitulo_frase': subtitulo_frase,
                'preco': preco,
                'area_privativa': area_privativa,
                'lista_detalhes': lista_resultado,
                'img_anunciante': img_anunciante,
                'nome_anunciante': nome_anunciante,
                'creci': creci,
                'pagina': pagina
                }
            # armazena o dicionário na lista
            dados_anuncio.append(dados)    
    except:
        print('erro inesperado ao extrair dados do anúncio')
    
    finally:
        return dados_anuncio

```

```python
url = 'https://www.dfimoveis.com.br/venda/df/brasilia/noroeste/imoveis?ordenamento=mais-recente'
soup = response_parser(url, headers)
lista = []


if soup:
    n_pagina = calcula_paginas(soup)

    for i in range(n_pagina):
        url_complemento = f'&pagina={i+1}'
        url_scrap = url + url_complemento  # monta url para scrap
        print(f'Processando página: {i+1}, URL: {url_scrap}')

        try:
            scrap = response_parser(url_scrap, headers)
            print()
            if soup:
                lista.extend(scrap_lista(scrap))
        except Exception as e:
            print(f'Erro na página {i+1}: {e}')
    
else: 
    print('Erro inicial ao acessar a página')
```
